```python
import subprocess
import anndata as ad
import scanpy as sc
from schelp.utils.config import load_config
from rich import print as rprint

paths = load_config(dataset_key="init")
subprocess.run(["tree", "-L", "1", paths["frozen"]])
```

```python
# Small balanced dataset with equal number of cells per supertype
adata_bal = sc.read(paths["frozen"] / "balanced_SEAAD_MTG_RNAseq_Singleome_final-nuclei.2024-06-18.h5ad", backed="r")
print(f"Number of cells: {adata_bal.shape[0]}")
print(f"Number of supertype categories: {adata_bal.obs['Supertype'].cat.categories.size}")
print(f"Number of supertype available: {adata_bal.obs['Supertype'].unique().size}")
print(f"Number of donors: {adata_bal.obs['Donor ID'].cat.categories.size}")
df = adata_bal.obs['Overall AD neuropathological Change'].value_counts().to_frame()
df["ratio"] = (df["count"] / df["count"].sum()).round(2)
display(df)
```
